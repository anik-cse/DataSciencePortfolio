```python
import pandas as pd
df = pd.read_csv('/content/heart failure.csv')
df
```

```python
y = df['DEATH_EVENT']
```

```python
y
```

```python
x=df.drop(['DEATH_EVENT'],axis=1)
```

```python
x
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x,y)
```

```python
tree.plot_tree(dt)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(50,40))
tree.plot_tree(dt,filled=True)
```

```python
x.columns
```

```python
tree.plot_tree(dt,filled=True,feature_names=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=.75,random_state=1)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
dt.fit(xtrain,ytrain)
```

```python
dt.score(xtrain,ytrain)
```

```python
dt.score(xtest,ytest)
```

```python
pred_y=dt.predict(xtest)
pred_y
```

```python
ytest
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, plot_roc_curve
```

```python
print(classification_report(ytest,pred_y))
```

```python
accuracy_score(ytest,pred_y)
```

```python
print(confusion_matrix(ytest,pred_y))
```

```python
import seaborn as sns
```

```python
cm=confusion_matrix(ytest,pred_y)
sns.heatmap(cm,annot=True)
```

```python
plot_roc_curve(dt,xtest,ytest)
```
