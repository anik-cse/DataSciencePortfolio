```python
# finding the outliers

def outliers(new_data, ft):
    Q1 = new_data[ft].quantile(0.25)
    Q3 = new_data[ft].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR 
    
    ls = new_data.index[ (new_data[ft] < lower_bound) | (new_data[ft] > upper_bound)]
    return ls
```

```python
new_data.rename(columns = {'fixed.acidity':'fixedacidity', 'volatile.acidity':'volatileacidity','citric.acid':'citricacid','residual.sugar':'residualsugar',
                      'chlorides':'chlorides','free.sulfur.dioxide':'freesulfurdioxide','total.sulfur.dioxide':'totalsulfurdioxide','density':'density',
                      'pH':'pH','alcohol':'alcohol','quality':'quality'}, inplace = True)

new_data.info()
```

```python

index_list = []
for feature in ['fixedacidity', 'volatileacidity','citricacid', 'residualsugar', 'chlorides', 'freesulfurdioxide', 'totalsulfurdioxide', 'density', 'pH','alcohol']:
    index_list.extend(outliers(new_data,feature))

index_list
```

```python
def remove(new_data, ls):
    ls = sorted(set(ls))
    new_data = new_data.drop(ls)
    return new_data
df_cleaned = remove(new_data, index_list)

df_cleaned.shape
```
